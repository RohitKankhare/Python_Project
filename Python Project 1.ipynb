# Drop Unnamed Colomn

```python
df = df.drop("Unnamed: 0", axis = 1)
print(df.head())
```

```python
df.head()
```

# Gender Distribution

```python
plt.figure(figsize= (5,5))
ax = sns.countplot(data = df, x = "Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

From the above chart we analysed that the number of females in the data is more than the number of male 

```python
gb = df.groupby("ParentEduc").agg({"MathScore": "mean", "ReadingScore": "mean", "WritingScore": "mean"})
print(gb)
```

```python
plt.figure(figsize= (4, 4))
sns.heatmap(gb, annot = True)
plt.title("Relationship between ParentEduc and StudentScores")
plt.show()
```

From the above chart we have conclude that the education of parents are more impactful on their kids exam scores

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore": "mean", "ReadingScore": "mean", "WritingScore": "mean"})
print(gb1)
```

```python
plt.figure(figsize= (4, 4))
sns.heatmap(gb1, annot = True)
plt.title("Relationship between ParentMaritalStatus and StudentScores")
plt.show()
```

From the above chart we have concluded that there is no/negligible impact on parent marital status

# Distribution of EthnicGroups

```python
groupA = df.loc[(df["EthnicGroup"] == "group A")].count()
groupB = df.loc[(df["EthnicGroup"] == "group B")].count()
groupC = df.loc[(df["EthnicGroup"] == "group C")].count()
groupD = df.loc[(df["EthnicGroup"] == "group D")].count()
groupE = df.loc[(df["EthnicGroup"] == "group E")].count()

l = ["groupA", "groupB", "groupC", "groupD", "groupE"]
mylist = [groupA["EthnicGroup"], groupB["EthnicGroup"], groupC["EthnicGroup"], groupD["EthnicGroup"], groupE["EthnicGroup"]]
print(mylist)
plt.pie(mylist, labels = l, autopct = "%1.2f%%")
plt.title("Distribution of Groups")
plt.show()
```
